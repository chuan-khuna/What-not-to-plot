```python
import pandas as pd
import numpy as np
import lorem
import matplotlib.pyplot as plt
from utils.data_generators import generate_pie
import seaborn as sns
from utils.plot import plot_split_pie
from matplotlib import patheffects

sns.set_theme(style='whitegrid')

FIGSIZE=(10, 6)
DPI=200
```

```python
df = pd.read_csv("./data/pie/normal/n16-20230424-2033.csv")
```

```python
fig = plt.figure(figsize=FIGSIZE, dpi=DPI)
plt.pie(df['value'], labels=df['long_name'], colors=['#355778'], wedgeprops={'linewidth': 0.5, 'edgecolor': 'white'})
plt.show()
```

```python
fig = plt.figure(figsize=FIGSIZE, dpi=DPI)

sns.barplot(df.sort_values('value', ascending=False), x='value', y='long_name', color='#355778')

# clean some element
sns.despine(bottom=True, left=True)
plt.xlabel("")
plt.ylabel("")
plt.grid(linewidth=0.05, axis='x', alpha=1.0, color='black')

plt.show()
```

```python
fig = plt.figure(figsize=(5, 10), dpi=DPI)

sns.barplot(df, x='value', y='long_name', color='#c9f1dd')


patheffect = patheffects.withStroke(linewidth=1, foreground='white', alpha=1.0)

for i, row in df.iterrows():
    plt.text(x=row['value']-0.01, y=i, 
             s=f"{row['long_name']}\n{row['value']}", ha='right', va='center', path_effects=[patheffect])
    

# clean some element
sns.despine(bottom=True, left=True)
plt.xlabel("")
plt.ylabel("")
plt.yticks([])
plt.grid(linewidth=0.05, axis='x', alpha=1.0, color='black')

plt.show()
```
